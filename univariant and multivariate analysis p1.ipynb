# Bivariant /Multivariant Analysis:Multivariate analysis refers to a set of statistical techniques used to examine data that involves more than one variable at a time. It helps uncover relationships, patterns, and dependencies between variables, which is especially useful when trying to understand complex datasets.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
titanic=pd.read_csv(r"C:\Users\user\Downloads\train.csv")
flights=sns.load_dataset('flights')
```

## Scatterplot(Numeric-numeric)

```python
tips.head()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## Bar plot( categorial-numerical)

```python
titanic.head()
```

```python
# A barplot is usefull for showing the average value of a numerical variable across different
# categories of a categorial variable
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
male_passenger=titanic[titanic['Sex']=='male']
male_passenger['Age'].mean()
```

```python
male_passenger=titanic[titanic['Sex']=='female']
male_passenger['Age'].mean()
```

```python
p1=titanic[(titanic['Pclass']==1)&(titanic['Sex']=='female')]
print(f' we have {p1.shape[0]} female passenger in Pclass 1')#without apply shape[0] it show row and column
```

```python
p1=titanic[(titanic['Pclass']==1)&(titanic['Sex']=='male')]
print(f' we have {p1.shape[0]} male passenger in Pclass 1')
```

```python
# Box plot(categorical-numerical)
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
sns.histplot(x=titanic[titanic['Survived']==0]['Age'],kde=True,stat='density')
sns.histplot(x=titanic[titanic['Survived']==1]['Age'],kde=True,stat='density')
```

```python
sns.kdeplot(x=titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(x=titanic[titanic['Survived']==1]['Age'])
```

```python
# HeatMap(Categorical-Categorical)
titanic.head()
```
